```python
def rotate(matrix, direction):
    if direction == 'right':
        return [list(row) for row in zip(*matrix[::-1])]
    elif direction == 'left':
        return [list(row) for row in zip(*matrix)][::-1]
    else:
        raise ValueError("direction must be 'left' or 'right'")
```

```python
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(data)
print(rotate(data, 'left'))
print(rotate(data, 'right'))
```

```python
class Book:
    def __init__(self, title, author, year, isbn, copies):
        self.title = title
        self.author = author
        self.year = year
        self.isbn = isbn
        self.copies = copies
        print(f"Book '{title}' added to catalog.")

    def change_copies(self, amount):
        """Updates availability: negative for checkout, positive for return."""
        if self.copies + amount >= 0:
            self.copies += amount
            return True
        print(f"Operation failed: Not enough copies of '{self.title}'.")
        return False

    def display_info(self):
        return f"'{self.title}' by {self.author} ({self.year}) | ISBN: {self.isbn} | Available: {self.copies}"

class User:
    def __init__(self, name, card_number):
        self.name = name
        self.card_number = card_number
        self.borrowed_books = []
        print(f"User {name} registered with card {card_number}.")

    def borrow_book(self, book):
        if book.change_copies(-1):
            self.borrowed_books.append(book)
            print(f"{self.name} successfully borrowed '{book.title}'.")
            return True
        return False

    def return_book(self, book):
        if book in self.borrowed_books:
            book.change_copies(1)
            self.borrowed_books.remove(book)
            print(f"{self.name} returned '{book.title}'.")
            return True
        print(f"Return failed: {self.name} does not have '{book.title}'.")
        return False

    def list_borrowed(self):
        return [book.title for book in self.borrowed_books]


class Library:
    def __init__(self):
        self.books = []
        self.users = []
        self.transaction_history = []
        print("Library Management System Initialized.")

    def add_book(self, book):
        self.books.append(book)

    def remove_book(self, book):
        if book in self.books:
            self.books.remove(book)

    def register_user(self, user):
        self.users.append(user)

    def delete_user(self, user):
        if user in self.users:
            self.users.remove(user)

    def issue_book(self, user, book):
        if user.borrow_book(book):
            self.transaction_history.append(f"ISSUED: '{book.title}' to {user.name}")
            return True
        return False

    def receive_return(self, user, book):
        if user.return_book(book):
            self.transaction_history.append(f"RETURNED: '{book.title}' by {user.name}")
            return True
        return False

    def search_books(self, query):
        """Searches by title or author."""
        results = [b.display_info() for b in self.books if query.lower() in b.title.lower() or query.lower() in b.author.lower()]
        return results if results else "No books found."

    def display_history(self):
        print("\n--- Library Transaction History ---")
        for record in self.transaction_history:
            print(record)
```

```python
library = Library()

book1 =Book('The Museum of Innocence', 'Orhan Pamuk', 2008, '9780571237029', 3)
book2 = Book('Sense and Sensibility', 'Jane Austen', 1811, '8596547395935', 2)
book3 = Book('The Book Thief', 'Markus Zusak', 2005, '0375842209', 1)

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

user1 = User('Wassilly Pupkin', 'A-0001')
user2 = User('Pavel Liubimov', 'A-0002')

library.register_user(user1)
library.register_user(user2)

library.issue_book(user1, book3)

library.issue_book(user2, book3)

library.issue_book(user2, book1)

library.receive_return(user1, book3)

library.issue_book(user2, book3)

user2.return_book(book3)

book3.change_copies(4)

book3.display_info()
















```
